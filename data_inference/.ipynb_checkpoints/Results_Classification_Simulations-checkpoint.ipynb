```python
import pickle
import itertools
import time
import os
from time import time

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from evaluation.classification import generate_class_results, get_classification_results, plot_conf_mat
```

# Results generation

```python
pickle_path = '/workspace/coniferas(1)/data_inference/pickles/simulations_no_fossil/'
res_path = "/workspace/coniferas(1)/data_inference/models/class/"

n_tips = ['674', '489', '87']

n_imgs_tested = 1000
results = dict()

for i in n_tips:
    with open(pickle_path + "dataset_" + i + "_10k.pkl", 'rb') as f:
        data = pickle.load(f)
    
    results[i] = dict()
        
    model_path = res_path + i + "_classification_"

    print('---', i, 'tips ---')
    results[i] = generate_class_results(model_path, data['X_test'][:n_imgs_tested],
                                        data['y_class_test'][:n_imgs_tested])
```

# Results summary

```python
get_classification_results(results)
```

# Confusion Matrices

```python
label_names = ["BD", "HE", "ME", "SAT", "SR", "WW"]
for i in results:
    plot_conf_mat((results[i]["y_pred"]), (results[i]["y_test"]), label_names, i)
```

# Train History

```python
def train_plot(ax, i, metric):
    ax.plot(results[i]['history'][metric],
            linestyle='-', label='Train', color='blue', linewidth=.7)
    ax.plot(results[i]['history']['val_' + metric],
            linestyle='--', label='Validation', color='blue', linewidth=.7)
    

    ax.set_xlabel('Epoch')
    ax.set_ylabel(metric)
    ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    
    ax.set_title(i)
```

```python
fig, ax = plt.subplots(1, len(n_tips), figsize=(40,7))
for i, j in enumerate(n_tips):
    train_plot(ax[i], j, 'accuracy')
    ax[i].grid(True)
plt.tight_layout(rect=[0, 0, 0.85, 1])
plt.show()
```

```python
fig, ax = plt.subplots(1, len(n_tips), figsize=(40,7))
for i, j in enumerate(n_tips):
    train_plot(ax[i], j, 'loss')
    ax[i].grid(True)
plt.tight_layout(rect=[0, 0, 0.85, 1])
plt.show()
```

## Results stratified
